```python
import sqlite3


# Program 2
# function to convert height from inches to feet
def convert_height(_height):
    return _height / 2.54 / 12


# function that convert weight from pounds to kilograms
def convert_weight(_weight):
    return _weight / 2.2


# Function to calculate ideal body weight
def ideal_weight(_weight, _height, _gender):
    _weight = convert_weight(_weight) if type(_weight) == int or type(_weight) == float else _weight
    _height = convert_height(_height) if type(_height) == int or type(_height) == float else _height
    if _gender == 'Male':
        return 50 + 2.3 * (_height - 5)
    elif _gender == 'Female':
        return 45.5 + 2.3 * (_height - 5)


# Function that determine if the user is over or under their IBW
def check_weight(_ideal_weight, _actual_weight):
    return 'underweight' if _actual_weight < _ideal_weight else 'overweight'


def get_recommendations(ideal__weight, actual__weight):
    if ideal__weight > actual__weight:
        return 'To reach your ideal weight, try to increase your food intake and do exercises that build up your'
    else:
        return ('To reach your ideal weight, try to reduce your food intake and do regular exercises that help '
                'to burn fat.')


# Function that stores user details and track their ideal body weight
def store_user_details(user_details):
    connection = sqlite3.connect('user_w.db')
    cursor = connection.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS users (
                        name TEXT,
                        height REAL,
                        weight REAL,
                        gender TEXT,
                        ibw REAL,
                        actual_weight REAL,
                        weight_status TEXT
                    )''')
    cursor.execute('''INSERT INTO users (name, height, weight, gender, ibw, actual_weight, weight_status)
                      VALUES (?, ?, ?, ?, ?, ?, ?)''', (user_details['name'], user_details['height'],
                                                        user_details['weight'], user_details['gender'],
                                                        user_details['ibw'], user_details['actual_weight'],
                                                        user_details['weight_status']))
    connection.commit()
    connection.close()


# Function to retrieve and display stored user details
def retrieve_user_details():
    connection = sqlite3.connect('user_details.db')
    cursor = connection.cursor()
    cursor.execute('''SELECT * FROM users''')
    users = cursor.fetchall()
    connection.close()
    if not users:
        print("No stored user details found.")
    else:
        print("Stored user details:")
        for user in users:
            print(f"Name: {user[0]}, Height: {user[1]}, Weight: {user[2]}, Gender: {user[3]}, "
                  f"Ideal Weight: {user[4]}, Actual Weight: {user[5]}, Weight Status: {user[6]}")


# Ask the user if they want to retrieve stored details
retrieve_choice = input("Do you want to retrieve stored weight details? (yes/no): ").lower()
if retrieve_choice == 'yes':
    retrieve_user_details()

# Get inputs from the user
name = input("Enter your name: ")
height = float(input("Enter your height in inches: "))
weight = float(input("Enter your weight in pounds: "))
gender = input("Enter your gender (Male/Female): ")

ideal_weight = ideal_weight(weight, height, gender)
actual_weight = float(input("Enter your actual weight in pounds: "))
weight_status = check_weight(ideal_weight, actual_weight)
recommendations = get_recommendations(ideal_weight, actual_weight)

print(f'Ideal Weight: {ideal_weight:.1f}kg')
print(f'Weight Status: {weight_status.capitalize()}')
print(f'Recommendations: {recommendations}')

user_details = {'name': name, 'height': height, 'weight': weight, 'gender': gender, 'ibw': ideal_weight,
                'actual_weight': actual_weight, 'weight_status': weight_status}
# Call the function to store weight details, including name
store_user_details(user_details)
```
